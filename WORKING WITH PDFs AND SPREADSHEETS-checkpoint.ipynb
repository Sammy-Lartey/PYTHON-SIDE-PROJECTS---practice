```python
f= open('Working_Business_Proposal.pdf','rb')  #open up the pdf file


pdf_text= []    #placeholder as a list

pdf_reader= PyPDF2.PdfFileReader(f)   #Create the reader object

for num in range(pdf_reader.numPages): 
    
    page= pdf_reader.getPage(num)    #getting the page at the particular number position
    
    pdf_text.append(page.extract_text())   #appending the extracted text(code  in bracket)
```

```python
import PyPDF2
```

```python
f= open('Find_the_Phone_Number.pdf','rb')
```

```python
pdf= PyPDF2.PdfReader(f)
```

```python
pdf.numPages
```

```python
import re       #using regular expressions
```

```python
pattern= r'\d{3}'
```

```python
all_text=''

for n in range (pdf.numPages):
    
    page= pdf.getPage(n)
    page_text= page.extract_text()
    
    all_text= all_text+' '+page_text
```

```python
all_text
```

```python
for match in re.finditer(pattern,all_text):
    print(match)
```
